```python
logpath = '../size_of-bak'
model_time=30
fields = ['p3_metallicity_radius']
data, tlen, tmax = get_data_dict(logpath)
dt = tmax // (tlen-1)
idx = int(model_time // dt) # index of time you want to model-stuck at final state for now
print('Max Time = %0.1f; T_len = %d;  DT = %0.1f; index = %d'%(tmax, tlen, dt, idx))
# print("PIDX's:")
# for k in data.keys():
#     print('\t%s'%k)
if idx >= tlen -1:
    idx = -1
print('Using index %d'%idx)
X = get_features(data, idx)
Y = get_labels(data, idx, fields)

```

```python
ffile = open('c_features.txt','w')
lfile = open('c_labels.txt','w')
for x,y in zip(X,Y):
    xw = ['%d'%(int(i)) for i in x]
    xw = ' '.join(xw)
    ffile.write('%s\n'%xw)
    lfile.write('%0.4f\n'%y)
ffile.close()
lfile.close()
```
